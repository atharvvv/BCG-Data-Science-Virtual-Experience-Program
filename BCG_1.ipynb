```python
import pandas as pd
# Load the data from CSV files into Pandas DataFrames (assuming you have already done this)
client_dataset = pd.read_csv('client_data.csv')
price_dataset = pd.read_csv('price_data.csv')
```

```python
# Check data types and handle missing values in client_data
print("Client Data Info:")
print(client_dataset.info())
print("\nNull Values in Client Data:")
print(client_dataset.isnull().sum())
```

```python
# Check data types and handle missing values in price_data
print("\nPrice Data Info:")
print(price_dataset.info())
print("\nNull Values in Price Data:")
print(price_dataset.isnull().sum())
```

```python
# Describe statistics for numerical variables in client_data
print("\nSummary Statistics for Numerical Variables in Client Data:")
print(client_dataset.describe())
```

```python
import pandas as pd
import matplotlib.pyplot as plt



# Describe statistics for numerical variables in client_data
numerical_variables = ['cons_12m', 'cons_gas_12m', 'cons_last_month', 'forecast_cons_12m',
                       'forecast_cons_year', 'forecast_discount_energy', 'forecast_meter_rent_12m',
                       'forecast_price_energy_off_peak', 'forecast_price_energy_peak', 'forecast_price_pow_off_peak',
                       'imp_cons', 'margin_gross_pow_ele', 'margin_net_pow_ele', 'nb_prod_act', 'net_margin',
                       'num_years_antig', 'pow_max']

# Descriptive statistics
print("Descriptive Statistics for Numerical Variables:")
print(client_dataset[numerical_variables].describe())

# Distribution plots for numerical variables
for var in numerical_variables:
    plt.figure(figsize=(8, 5))
    plt.hist(client_dataset[var], bins=30, edgecolor='k')
    plt.xlabel(var)
    plt.ylabel('Frequency')
    plt.title(f'Distribution of {var}')
    plt.show()
```

```python
# Analyze churn variable
churn_proportion = client_dataset['churn'].value_counts(normalize=True)

# Display the proportion of churned and non-churned clients
print("Proportion of Churned and Non-Churned Clients:")
print(churn_proportion)
```

```python
# Create a bar plot for churn proportion
plt.figure(figsize=(6, 4))
plt.bar(churn_proportion.index, churn_proportion.values)
plt.xticks(churn_proportion.index, ['Non-Churned', 'Churned'])
plt.xlabel('Churn Status')
plt.ylabel('Proportion')
plt.title('Proportion of Churned and Non-Churned Clients')
plt.show()
```
