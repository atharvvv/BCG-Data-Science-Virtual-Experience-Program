```python
import nltk
from nltk.corpus import stopwords
from string import punctuation
from prettytable import PrettyTable
from gensim.models import Word2Vec  # Add this line
from sklearn.metrics.pairwise import cosine_similarity

# ... (rest of your code)

# Download NLTK data
nltk.download('punkt')
nltk.download('stopwords')

# List of 10 sentences
sentences = [
    "Superman, also known as Clark Kent, is the last son of Krypton and possesses superhuman strength and the ability to fly.",
    "Batman, a.k.a. Bruce Wayne, fights crime in Gotham City using his intellect, martial arts skills, and an array of gadgets.",
    "Wonder Woman, or Diana Prince, is an Amazonian princess with superhuman strength, agility, and a magical lasso.",
    "The Flash, also known as Barry Allen, is the fastest man alive, capable of moving at incredible speeds.",
    "Aquaman, a.k.a. Arthur Curry, rules the underwater kingdom of Atlantis and commands the power of the seas.",
    "Green Lanterns, members of the Green Lantern Corps, wield power rings fueled by willpower to protect the universe.",
    "Harley Quinn, formerly known as Dr. Harleen Quinzel, is a chaotic and unpredictable villain often associated with the Joker.",
    "The Joker, Batman's arch-nemesis, is a deranged criminal mastermind known for his chaotic and sadistic nature.",
    "Supergirl, or Kara Zor-El, is Superman's cousin with similar powers, including super strength and flight.",
    "The Arrow, a.k.a. Oliver Queen or Green Arrow, is a skilled archer and vigilante protecting Star City."
]


```

```python
# Function to perform POS tagging and create a table
def pos_tagging_table(sentence, statement_number, table):
    stop_words = set(stopwords.words('english'))
    tokens = nltk.word_tokenize(sentence)
    filtered_tokens = [word.lower() for word in tokens if word.isalpha() and word.lower() not in stop_words]

    pos_tags = nltk.pos_tag(filtered_tokens)

    # Add data to the table
    table.add_row([
        f"Statement {statement_number}",
        sentence,
        "\n".join([f"{token}" for token in filtered_tokens]),
        "\n".join([f"{pos}" for _, pos in pos_tags])
    ])

    return filtered_tokens

# Create a PrettyTable for POS tagging
result_table_pos = PrettyTable()
result_table_pos.field_names = ["Statement Number", "Statement", "Tokens", "POS"]

# Collect all tokens for POS tagging
all_tokens_pos = []

# Perform POS tagging for each sentence and add to the table
for i, sentence in enumerate(sentences, 1):
    tokens = pos_tagging_table(sentence, i, result_table_pos)
    all_tokens_pos.extend(tokens)

# Train Word2Vec model for POS tagging
model_pos = Word2Vec([all_tokens_pos], vector_size=100, window=5, min_count=1, workers=4)

# Create a PrettyTable for similarity scores
result_table_similarity = PrettyTable()
result_table_similarity.field_names = ["Statement Number 1", "Token 1", "Statement Number 2", "Token 2", "Similarity"]

# Function to calculate cosine similarity between two tokens
def calculate_similarity(token1, token2, model):
    similarity = cosine_similarity([model.wv[token1]], [model.wv[token2]])[0][0]
    return similarity

# Calculate similarity scores
similarity_scores = []
for i in range(len(all_tokens_pos)):
    for j in range(i + 1, len(all_tokens_pos)):
        statement_number_i = all_tokens_pos.index(all_tokens_pos[i]) // len(sentences) + 1
        statement_number_j = all_tokens_pos.index(all_tokens_pos[j]) // len(sentences) + 1

        similarity = calculate_similarity(all_tokens_pos[i], all_tokens_pos[j], model_pos)
        similarity_scores.append((statement_number_i, all_tokens_pos[i], statement_number_j, all_tokens_pos[j], f"{similarity:.4f}"))

# Add similarity scores to the table
for score in similarity_scores:
    result_table_similarity.add_row([score[0], score[1], score[2], score[3], score[4]])

# Print the POS tagging table
print("\nPOS Tagging Table:")
print(result_table_pos)

# Print the similarity scores table
print("\nSimilarity Scores:")
print(result_table_similarity)
```

```python
# Function to perform POS tagging and create a table
def pos_tagging_table(sentence, statement_number, table):
    tokens = nltk.word_tokenize(sentence)
    pos_tags = nltk.pos_tag(tokens)

    # Add data to the table
    table.add_row([
        f"Statement {statement_number}",
        sentence,
        "\n".join([f"{token}" for token, _ in pos_tags]),
        "\n".join([f"{pos}" for _, pos in pos_tags])
    ])

# Create a PrettyTable
result_table = PrettyTable()
result_table.field_names = ["Statement Number", "Statement", "Tokens", "POS"]

# Perform POS tagging for each sentence and add to the table
for i, sentence in enumerate(sentences, 1):
    pos_tagging_table(sentence, i, result_table)

# Print the table
print(result_table)
```
