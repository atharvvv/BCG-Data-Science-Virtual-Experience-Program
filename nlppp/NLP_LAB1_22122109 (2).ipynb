##Importing Text File

```python
with open('nlp.txt' , 'r' , encoding = 'utf-8') as f:
  lines = f.read()
  print(lines)
```

##Define a list of stop words

```python
# Define a list of stop words
stop_words = set(spacy.lang.en.stop_words.STOP_WORDS)
```

##Creating Tokens , remove symbols and number ,  describe the POS , remvoing  prefixes and suffixes and describtion



###*  token.is_alpha: It checks if the current token contains only alphabetic characters, excluding symbols, numbers, and special characters.
###*   token.text.lower() not in stop_words: It checks if the lowercase version of the token (to make the comparison case-insensitive) is not in the set of stop words defined earlier.
###* Creating Tabel of the output



```python
#import spacy and Load the Langauage Library
import spacy
nlp = spacy.load('en_core_web_sm') #storing the english dictionary as an object named 'nlp'

#Create a Doc Object
doc = nlp(lines)

# Create a list to store token information
token_info = []

# Print each token separately, excluding symbols and numbers
for token in doc:
    if token.is_alpha and token.text.lower() not in stop_words:
        token_info.append([token.text, token.pos_, token.dep_, token.lemma_ , token.ent_type_])

# Include named entities in the table
for ent in doc.ents:
    token_info.append([ent.text, ent.label_, "", "", ""])
```

```python
# Define headers for the table
headers = ["Text", "POS", "Dependency", "Lemma", "Entity Type"]

# Print the table using tabulate
print(tabulate(token_info, headers=headers, tablefmt="grid"))
```
