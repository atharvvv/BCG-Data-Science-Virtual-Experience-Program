```python
from PyPDF2 import PdfReader
read = PdfReader('nlp_lab2.pdf')
page = read.pages[0]
lines = page.extract_text()
print(lines)
```

```python
stop_words = nlp.Defaults.stop_words
```

```python
doc = nlp(lines)
token_info = []

for tokens in doc:
  if(tokens.is_alpha and tokens.text.lower() not in stop_words):
    token_info.append(tokens.text)
print()
```

```python
def get_synonyms(word):
  synonyms = []
  for syn in wordnet.synsets(word):
    for lemma in syn.lemmas():
      synonyms.append(lemma.name())
  return synonyms
```

```python
def get_antonyms(word):
  antonyms = []
  for syn in wordnet.synsets(word):
      for lemma in syn.lemmas():
        for antonym in lemma.antonyms():
          antonyms.append(antonym.name())
  return(antonyms)



```

```python
for tokens in token_info:
  synonyms = get_synonyms(tokens)
  antonyms = get_antonyms(tokens)
  synonyms = list(set(synonyms))
  antonyms = list(set(antonyms))


  synonyms.sort()
  antonyms.sort()

  print(f"token : {tokens}")
  print(f"synonyms : {synonyms}")
  print(f"antonyms : {antonyms}")



```
