```python
import nltk
from nltk.tokenize import word_tokenize
nltk.download('punkt')
# Tokenize the text into words
tokens = word_tokenize(text)
# Print the tokens
print("Tokens:", tokens)
```

```python
from tabulate import tabulate
```

```python
doc = nlp(text)



# Initialize lists to store token information and word frequencies
token_info = []
word_frequencies = Counter()

# Iterate over tokens in the processed text
for token in doc:
    if token.is_alpha and token.text.lower() not in nlp.Defaults.stop_words:
        # Append token information (text, POS, dependency, lemma, entity type) to token_info
        token_info.append([token.text, token.pos_, token.dep_, token.lemma_, token.ent_type_])

# Append entity type to token_info for named entities
for ent in doc.ents:
    token_info.append([ent.text, "", "", "", ent.label_])

# Count the frequency of each word
for token in doc:
    if token.is_alpha and token.text.lower() not in nlp.Defaults.stop_words:
        word_frequencies[token.lemma_] += 1

# Append word frequency to token_info
for row in token_info:
    row.append(word_frequencies.get(row[3], 0))  # Append the frequency (lemma) to each row

# Define headers for the table
headers = ["Text", "POS", "Dependency", "Lemma", "Entity Type", "Frequency"]

# Print the table using tabulate
print("Token Information with Frequency:")
print(tabulate(token_info, headers=headers, tablefmt="grid"))

# Extract the top 15 words and their frequencies
top_words = [item[0] for item in word_frequencies.most_common(15)]
top_frequencies = [word_frequencies[word] for word in top_words]

# Plot the bar graph for top 15 words
plt.figure(figsize=(12, 6))
plt.bar(top_words, top_frequencies)
plt.xlabel('Words')
plt.ylabel('Frequency')
plt.title('Top 15 Words and Their Frequencies')
plt.xticks(rotation=45)
plt.show()
```

```python
# Process the text using spaCy
doc = nlp(text)

# Initialize lists to store token information and word frequencies
token_info = []
word_frequencies = Counter()

# Get stop words for Spanish
stop_words = spacy.lang.es.stop_words.STOP_WORDS

# Iterate over tokens in the processed text
for token in doc:
    if token.is_alpha and token.text.lower() not in stop_words:
        # Append token information (text, POS, dependency, lemma, entity type) to token_info
        token_info.append([token.text, token.pos_, token.dep_, token.lemma_, token.ent_type_])

# Append entity type to token_info for named entities
for ent in doc.ents:
    token_info.append([ent.text, "", "", "", ent.label_])
```

```python
# Count the frequency of each word
for token in doc:
    if token.is_alpha and token.text.lower() not in stop_words:
        word_frequencies[token.lemma_] += 1

# Append word frequency to token_info
for row in token_info:
    row.append(word_frequencies.get(row[3], 0))  # Append the frequency (lemma) to each row

# Define headers for the table
headers = ["Text", "POS", "Dependency", "Lemma", "Entity Type", "Frequency"]
```

```python
# Print the table using tabulate
print("Token Information with Frequency:")
print(tabulate(token_info, headers=headers, tablefmt="grid"))
```
